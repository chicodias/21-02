```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
import yfinance as yf

bvsp= yf.Ticker("^BVSP")
bvsp
```

```python
aapl_historical = bvsp.history(start="2020-06-01", end="2021-06-11", interval="1d")
aapl_historical.head()
```

```python
# plot the data and MA
import plotly.express as px
fig = px.line(aapl_historical, x=aapl_historical.index , y="Close", template = 'plotly_dark')
fig.show()
```

```python

# rolling 5-day moving average
aapl_historical['ma5'] = aapl_historical['Close'].rolling(window=5).mean()

# drop first five observations
aapl_historical = aapl_historical.iloc[5:]

```

```python

#defines start = "2020-06-08" and end = "2021-06-11" dates

start = dt.datetime(2020, 6, 8)

end = dt.datetime(2021, 6, 11)
```

```python
# index with 'D' frequency
index = pd.date_range(start=start, end=end, freq='D')
```

```python
aapl_historical.index
```

```python
# take only aapl_historical in new index    
aapl_historical2 = aapl_historical.reindex(index, method='nearest')
```

```python
aapl_historical2.head(10)
```

```python
#convert aapl_historical to a pd.Series with business days as index
aapl_series = pd.Series(aapl_historical2["ma5"], index=index)
```

```python
exp.plot_model(plot="acf")

# NOTE: you can customize the plots with kwargs - e.g. number of lags, figure size (width, height), etc
# data_kwargs such as `nlags` are passed to the underlying functon that gets the ACF values
# figure kwargs such as `fig_size` & `fig_template` are passed to plotly and can have any value that plotly accepts
exp.plot_model(plot="pacf", data_kwargs={'nlags':36, }, fig_kwargs={'fig_size': [800, 500], 'fig_template': 'simple_white'})
```
